```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import os



title1 = []
main1 = []
new_url1 = []
recovery1 = []
idds = []
createTime1 = []

url = 'https://www.myhousing.com.tw/index.php?option=com_kunena&view=search&Itemid=539&query=%E5%8F%B0%E5%8C%97&searchdate=all&childforums=1&catids=8&limitstart=0'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36',
    'Referer': 'https://www.myhousing.com.tw/index.php?option=com_kunena&view=search&Itemid=539',
    'Host': 'www.myhousing.com.tw'
}

for i in range(111):
    req = requests.get(url, headers=headers)
    soup = BeautifulSoup(req.text, 'html.parser')
    a_tags = soup.select('div[class="badger-left badger-info khistory"]')
    
 
    
    for tag in a_tags:
        idd = tag.select_one('a[class="hasTooltip"]')['href'].split('&')[3].split('=')[1]
#             print(idd)

        if idd not in idds:
            recoverysinside = []


            idds.append(idd)

            ch_url = 'https://www.myhousing.com.tw/' + tag.select_one('a[class="hasTooltip"]')['href']
            new_url = ch_url.replace(ch_url.split('=')[5], '0')
            new_url1.append(new_url)

            new_req = requests.get(new_url, headers=headers)
            new_soup = BeautifulSoup(new_req.text, 'html.parser')
            new_tag = new_soup.select('small[class="text-muted pull-right"]')
            datetime = [x.select_one('span[class="kdate"]').text for x in new_tag]
            create_time = datetime[0] + ':00'
            createTime1.append(create_time)
            

            title = tag.select_one('a[class="hasTooltip"]').text
            title1.append(title)

            main = tag.select_one('a[class="hasTooltip"]')['title']
            main1.append(main)

            recovery = tag.select_one('div[class="kmessage"]').text.replace('\t','').replace('\n','')
            recoverysinside.append(recovery)
            recovery1.append(recoverysinside)
            

        else:
            recovery = tag.select_one('div[class="kmessage"]').text.replace('\t','').replace('\n','')
            recovery1[idds.index(idd)].append(recovery)
            
#         print(new_url)
#         print(title)
#         print(main)
#         print(recovery)  
    
    pages = soup.select('div[class="pagination hidden-phone"]')
    for page in pages:
        url = 'https://www.myhousing.com.tw/' + page.select('a[class="hasTooltip"]')[-2]['href']
    time.sleep(5)
#     print(df)
columns=['url', 'title', 'main', 'recovery', 'createTime']
df = pd.DataFrame(columns = columns)

df['url'] = new_url1
df['title'] = title1
df['main'] = main1
df['recovery'] = recovery1
df['createTime'] = createTime1
print(df)
df.to_csv('./myhousing.csv', index=False, encoding='utf-8-sig')
```
